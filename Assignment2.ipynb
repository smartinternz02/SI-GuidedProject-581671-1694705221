```python
sns.displot(df["Distance from the airport"])
```

```python
x = df.head(20)["Postal Code"].unique()
#print(x)
#count of each unique value of postal code
y= list()
for i in x :
  y.append(df.head(20)["Postal Code"].value_counts()[i])
#print(y)

plt.pie(y,labels=x)
plt.show()

#from the pie chart  we can see that in first 20 observations, the Postal code 122006
#has highest number of houses.
```
