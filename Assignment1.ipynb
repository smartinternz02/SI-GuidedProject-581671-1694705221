```python
data={
    "name":["AMAN","RAMAN","NAMAN","AMIT"],
    "roll number":['21bce1303','21bce1403','20bme1310','21bce1310'],
    "age": [18,17,19,20],
    "total marks":[450,np.nan,470,460]
}
```

```python
# creating dataframe with name df with 4 features and 4 observations
df = pd.DataFrame(data)
print(df)
```

```python
#renaming the column names.
print("column name before making changes : ",end=" ")
print(df.columns.values.tolist())
df.rename(columns = {'name':'Random value 1',
                     "roll number":'Random value 2',
                     "age":'Random value 3',
                     "total marks":'Random value 4'}, inplace = True)
print("column name after making changes : ",end=" ")
print(df.columns.values.tolist())
```

```python
#finding descriptive analysis of the dataframe
print(df.describe())
```

```python
#finding null vales in the dataframe
print(df.isnull())
print()
print()
#finding data type of all the columns in dataframe
print(df.dtypes)
```

```python
#display the Random value 2 and Random value 3 columns with loc and iloc method
print(df.loc[:,["Random value 2", "Random value 3"]])
print(df.iloc[:,[1,2]])
```
